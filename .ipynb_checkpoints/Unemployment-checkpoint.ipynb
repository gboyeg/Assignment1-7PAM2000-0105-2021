```python
# Importing libraraies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing Datasets
data = pd.read_csv("unemployment_data_us.csv")
data.head()
```

```python
# Checking the information of the dataset
data.info()
```

```python
# Checking the number of missing data in each columns
data.isnull().sum()
```

```python
# Filling the missing data using the mean values of each columns
data.fillna(data.mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# Checking the count of unique data
data.nunique()
```

```python
# CHecking Value count for all the column in th dataset
for column in data.columns:
    print(data[column].value_counts().head())
```

```python
# Checking the summary of the dataset
data.describe()
```

```python
# Scattered plot od the Men  vs  Women
plt.figure(figsize=(10,8))
plt.scatter(x=data['Men'], y=data['Women'], c ="red", marker='*')
 
# To show the plot
plt.show()
```

```python
# Plotting pie chat of the Year column
plt.figure(figsize = (10,8))
data['Year'].value_counts().plot(kind = 'pie', autopct = '%.2f%%')
plt.show()
```

```python
# Plotting bar chat count of the Year column
plt.figure(figsize = (10,8))
sns.countplot(x = 'Year', data = data)
plt.show()
```

```python
# multi-line plotting based on race

plt.figure(figsize=(20,10))
plt.plot( 'Date', 'White', data=data, color='magenta')
plt.plot( 'Date', 'Black', data=data, color='red')
plt.plot( 'Date', 'Asian', data=data, color='blue')
plt.plot( 'Date', 'Hispanic', data=data, color='yellow')
plt.legend()
```

```python
# multi-line plotting based on education

plt.figure(figsize=(20,10))
plt.plot( 'Date', 'Primary_School', data=data, color='red')
plt.plot( 'Date', 'High_School', data=data, color='yellow')
plt.plot( 'Date', 'Associates_Degree', data=data, color='magenta')
plt.plot( 'Date', 'Professional_Degree', data=data, color='blue')
plt.legend()
```
